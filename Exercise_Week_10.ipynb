# Get the set of product that has been purchased


Get the unique product that has been purchased

```python
unique_products = df["6"].unique()

print(set(unique_products))
```

## Preprocess Data

In this step, we will transform our dataset so that we will have a one hot encoding based on the purchased products.

```python
#create an itemset based on the products
itemsets = []
for index, row in df.iterrows():
    itemset = set(row)
    itemsets.append(itemset)

print("Itemsets:")
for i, itemset in enumerate(itemsets, start=1):
    print(f"Transaction {i}: {itemset}")


# encoding the feature
one_hot_encoded = pd.DataFrame(0, columns=unique_products, index=range(1, len(itemsets) + 1))

for i, itemset in enumerate(itemsets, start=1):
    one_hot_encoded.loc[i, itemset] = 1

print("\nOne-Hot Encoded DataFrame:")
print(one_hot_encoded)
```

```python
  # create new dataframe from the encoded features
new_df_from_encoded = one_hot_encoded.copy()

  # show the new dataframe
new_df_from_encoded.head()
```

Since, the encoded dataframe consist of the empty column. We will drop the NaN column or select all columns other than the first column.

```python
new_df_from_encoded = new_df_from_encoded.loc[:, new_df_from_encoded.columns.notna()]

new_df_from_encoded.head()
```

## Apriori Algorithm

We will use appriori algorithm to determine the frequently purchased products. 
For this case study, we will min_support=0.2

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(new_df_from_encoded, min_support=0.2, use_colnames=True)
print(frequent_itemsets)
```

Then, we will generate association rule of the frequent itemset based on confidence level with the threshold=0.6

```python
rules_confidence_06 = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

print(rules_confidence_06)
```

Provide explanation about __antecedent support__, __consequent support__, __support__, __confidence__, __lift__, __leverage__ and __conviction__

Antecedent Support:
- Definition: The support of the antecedent itemset A.
- Formula: support(A)=proportion of transaction containing A.
- Interpretation: Measures the frequency of occurrence of the antecedent itemset A in the dataset.

Consequent Support:
- Definition: The support of the consequent itemset C.
- Formula: support(C)=proportion of transaction containing C.
- Interpretation: Measures the frequency of occurrence of the antecedent itemset C in the dataset.

Support:
- Definition: The support of the combined itemset A∪C
- Formula: support(A→C)=support(A∪C).
- Interpretation: Measures the frequency of occurrence of the combined itemset A∪C in the dataset.

Confidence:
- Definition: The conditional probability of observing the consequent C given the antecedent A.
- Formula: confidence(A→C)=support(A→C)/support(A).
- Interpretation: Measures how often the rule A→C is true.

Lift:
- Definition: Compares the likelihood of observing A and C together against the likelihood if they were independent.
- Formula: lift(A→C)=confidence(A→C)/support(C).
- Interpretation: A lift value greater than 1 indicates a positive correlation between A and C.

Leverage:
- Definition: Measures the difference between the observed frequency of A and C appearing together and the frequency expected under independence.
- Formula: levarage(A→C)=support(A→C)−support(A)×support(C).
- Interpretation: A value of 0 indicates independence.

Conviction:
- Definition: Measures how much the consequent C depends on the antecedent A.
- Formula: conviction(A→C)=(1−support(C))/(1−confidence(A→C)).
- Interpretation: A high conviction value indicates a strong dependency.
